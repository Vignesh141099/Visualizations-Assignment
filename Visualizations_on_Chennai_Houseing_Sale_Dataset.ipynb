```python
sns.distplot(chs['REG_FEE'],kde=False,color= 'r')
```

```python
sns.rugplot(chs['REG_FEE'],color='orange')
```

```python
sns.kdeplot(chs['N_BEDROOM'])
```

```python

sns.barplot(x='STREET',y='COMMIS',data=chs,color='violet')
```

```python
plt.figure(figsize=(12,3))
sns.boxplot(y='REG_FEE',x='AREA',data=chs,palette='Set2')
```

```python
sns.violinplot(x='STREET',y='REG_FEE',data=chs,palette='coolwarm')
```

```python
sns.stripplot(x='STREET',y='SALES_PRICE',data=chs,palette='rainbow')
```

```python
sns.swarmplot(x='STREET',y='REG_FEE',data=chs,palette='flare')
```

```python
sns.catplot(x='STREET',y='COMMIS',data=chs,palette='autumn')
```
