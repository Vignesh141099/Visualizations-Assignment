```python
data = data[~((data.radius_mean<(q1-1.5*os))|(data.radius_mean>(q3+1.5*os)))]
```

```python
def stats(y):
  sum = 0
  sum2 = 0
  items = 0
  for x in y:
    sum += x
    sum2 += x**2
    items +=1
  mean = sum/items
  variance = ((sum2/items) - mean**2) 
  return mean, variance
```

```python
stats([10,15,22,0,15,82])
```

```python
class Fibo:
  def __iter__(fibo):
    fibo.first = -1
    fibo.second = 1
    return fibo
  def __next__(fibo):
    fibo.third = fibo.first + fibo.second # 0
    fibo.first = fibo.second # 
    fibo.second = fibo.third

    return fibo.third
```

```python
i = iter(Fibo())
for j in range(10):
  print(next(i))
```
